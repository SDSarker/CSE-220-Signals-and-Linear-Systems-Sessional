```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xstart=0
xend=1.1
increment=0.1

x = np.arange(xstart, xend, increment)
y = np.power(x, 2)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot of y = x^2')
plt.fill_between(x,y, color='skyblue', alpha=0.5)
plt.show()
```

```python
#Integration
a=0
b=1
N=10

x=np.linspace(a,b,N+1)
y=np.power(x,2)

y_right=y[1:]
y_left=y[:-1]

dx=(b-a)/N
A=(dx/2)*np.sum(y_right + y_left)

print("The approximate value of the integral is:", A)
plt.plot(x, y)
plt.fill_between(x,y,color='lightgreen', alpha=0.5)
plt.show()
```

```python
from scipy import integrate

a=0
b=1

def f(x):
    return x**2

A,err=integrate.quad(f,a,b)
```

```python
print("The approximate value of the integral is:", A)
print("Estimated error in the integral is:", err)
```

```python
import numpy.polynomial.polynomial as poly

p=[3,-1,2,1]
#function is x^3 + 2*x^2 -x + 3

I=poly.polyint(p)
print("The integral of the polynomial is:", I)

a=-1
b=1
A=poly.polyval(b,I)-poly.polyval(a,I)
print("The definite integral from", a, "to", b, "is:", A)
```
