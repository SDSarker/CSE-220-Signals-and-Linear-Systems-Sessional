```python
# unique + inverse + bincount + argmax (label grouping pattern)
labels = np.array(list("aabca"))
dur = np.array([3, 5, 2, 4, 6])
print("labels=",labels)
unique_lbl, inv = np.unique(labels, return_inverse=True)
print("unique_lbl = ",unique_lbl)
print("inv = ",inv)
print("dur=",dur)

total_dur = np.bincount(inv, weights=dur).astype(np.int64)
top_i = np.argmax(total_dur)

print("total_dur = ",total_dur)
print("top_i = ",top_i)

unique_lbl, total_dur, (unique_lbl[top_i], total_dur[top_i])
```

```python
def time_shift_signal(x : np.ndarray, k : int) -> np.ndarray:
    return np.roll(x,k)
```

```python
def time_scale_signal(x : np.ndarray, k : int) -> np.ndarray:
    temp=np.zeros_like(x)
    second_half=np.array(x[8::k])
    x=np.flip(x)
    x=np.array(x[8+k::k])
    x=np.flip(x)
    temp[8:8+np.size(second_half)]+=second_half
    temp[8-np.size(x):8]+=x
    return temp
```

## 3) Matplotlib (plots + formatting + subplots + save)

```python
import matplotlib.pyplot as plt
```

```python
# basic line plot
X = np.linspace(0, 10, 40)
Y = X**2

plt.figure(figsize=(6, 4))
plt.plot(X, Y, label="quadratic")
plt.plot(X, X, label="linear")
plt.grid()
plt.title("Demo")
plt.xlabel("X Axis")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# sine wave
t = np.linspace(0, 10*np.pi, 1000)
plt.figure(figsize=(6, 3))
plt.plot(t, np.sin(t))
plt.show()
```

```python
# stem plot + axis lines
n = np.arange(9)
x = np.array([0, 2, -3, 0, 3, 5, 4, -4, -5])

plt.figure(figsize=(6, 4))
plt.stem(n, x, basefmt=" ")
plt.axhline(y=0, color="black", linewidth=1)
plt.axvline(x=0, color="black", linewidth=1)
plt.xlabel("t")
plt.ylabel("func(t)")
plt.grid(True, alpha=0.3)
plt.title("Discrete-Time Signal")
plt.show()
```

```python
# image + contour
C = np.random.random((8, 8))

plt.figure(figsize=(4, 4))
plt.imshow(C)
plt.show()

plt.figure(figsize=(4, 4))
plt.contour(C)
plt.show()

plt.figure(figsize=(4, 4))
plt.contourf(C)
plt.show()
```

```python
# subplots + axes methods + suptitle + tight_layout + savefig
fig, axs = plt.subplots(2, 2, figsize=(6, 6))
axs[0, 0].imshow(C)
axs[0, 0].set_title("imshow")

axs[0, 1].stem(n, x, basefmt=" ")
axs[0, 1].axhline(y=0, color="black", linewidth=1)
axs[0, 1].set_title("stem")

axs[1, 0].plot(X, Y)
axs[1, 0].set_title("plot")

axs[1, 1].boxplot(rng.normal(0, 1, 200))
axs[1, 1].set_title("boxplot")

fig.suptitle("Subplots")
plt.tight_layout()
plt.savefig("fig.png")
plt.show()
```

```python
# bar + stacked bar (Axes API + layout='constrained')
months = ["Jan", "Feb", "Mar", "Apr"]
a = np.array([23, 25, 28, 30])
b = np.array([30, 32, 31, 33])
c = np.array([15, 18, 22, 25])

fig, axs = plt.subplots(2, 2, figsize=(8, 6), layout="constrained")

axs[0, 0].plot(months, a, label="A")
axs[0, 0].plot(months, b, label="B")
axs[0, 0].plot(months, c, label="C")
axs[0, 0].set_title("Sales")
axs[0, 0].legend()
axs[0, 0].grid(True)

bars = axs[0, 1].bar(["A", "B", "C"], [a[-1], b[-1], c[-1]])
bars[0].set_color("blue")
axs[0, 1].grid(True, axis="y")
axs[0, 1].set_title("Last Month")

axs[1, 0].scatter(months, a, s=50)
axs[1, 0].set_title("Scatter")
axs[1, 0].grid(True)

q = np.array(["H1", "H2"])
qa, qb, qc = a.reshape(2, 2).sum(axis=1), b.reshape(2, 2).sum(axis=1), c.reshape(2, 2).sum(axis=1)
axs[1, 1].bar(q, qa, label="A")
axs[1, 1].bar(q, qb, bottom=qa, label="B")
axs[1, 1].bar(q, qc, bottom=qa+qb, label="C")
axs[1, 1].set_title("Stacked")
axs[1, 1].legend()
axs[1, 1].grid(True, axis="y")

fig.suptitle("Mini Dashboard")
plt.show()
```
