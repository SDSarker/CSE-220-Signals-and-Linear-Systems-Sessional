```python
import numpy as np
```

```python
class Student:
    uni= "BUET"
    def __init__(self,name,id,*grades):
        self.name=name
        self.id=id
        self.grades=np.array(grades)
    def add_grades(self,grade):
        self.grades= np.append(self.grades,grade)
    def get_average(self):
        return self.grades.mean()
    def get_letter_grade(self):
        x=self.get_average()
        if x>=4:
            return 'A+'
        elif x>=3.75:
            return 'A'
        elif x>=3.5:
            return 'A-'
        else:
            return 'B'
```

```python
a=Student("Soumya",142,4,3,4,3)
b=Student("Arjit",138,2,3,4,2,2)
c=Student("Bhuban",136,4,4,4,4,4)

a.add_grades(4)
print(f'{a.get_average()} ---> {a.get_letter_grade()}')
print(f'{b.get_average()} ---> {b.get_letter_grade()}')
print(f'{c.get_average()} ---> {c.get_letter_grade()}')
```
