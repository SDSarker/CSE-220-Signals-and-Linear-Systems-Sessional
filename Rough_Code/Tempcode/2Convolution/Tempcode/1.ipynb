```python
import numpy as np

n=int(input())
ls=[]

for i in range(0,n,1):
    a,b,c=map(str,input().split(" "))
    
    label= a
    amp=int(b)
    dur=int(c)
    ls.append((label,amp,dur))
```

```python
ls
```

```python
lbl=[]
amp=[]
dur=[]
for i in ls:
    lbl.append(i[0])
    amp.append(i[1])
    dur.append(i[2])
```

```python
lb=np.array(lbl,dtype=str)
am=np.array(amp,dtype=np.int64)
du=np.array(dur,dtype=np.int64)

unique_lbl,inv= np.unique(lb,return_inverse=True)
total_dur= np.bincount(inv,weights=du)
total_energy=np.bincount(inv,weights=am*am*du)


for i in range (np.size(unique_lbl)):
    print(f'{unique_lbl[i]} {int(total_dur[i])} {int(total_energy[i])}')

print(f'TOP {unique_lbl[total_energy.argmax()]} {total_energy.max()}')
```
