```python
data = """\
4 3
70 45 80
90 75 30
40 60 20
100 50 90
100 75 50
2
"""
```

```python
import numpy as np
import sys
```

```python
#for input from console
# tok=sys.stdin.read().split()
# it=iter(tok)

tok=data.split()
it=iter(tok)
```

```python
S=int(next(it))
M=int(next(it))


A=np.zeros((S,M),dtype=np.int64)
for i in range(S):
    for j in range(M):
        A[i,j]=int(next(it))

max_marks=np.zeros(M)
for i in range(M):
    max_marks[i]=int(next(it))
```

```python
Normalized=A/max_marks * 100
print("Normalized Matrix : ")
print(Normalized)
```

```python
print("Student Summary")

Avg=A.mean(axis=1)
Best=A.max(axis=1)
Best_code=A.argmax(axis=1)
for i in range(S):
    print(f'Roll : {i} --> Best: {Best[i]} in Subject:{Best_code[i]}  Average: {Avg[i]}')
```

```python
print("Best K: ")
avg=A.mean(axis=1)
roll=np.arange(0,S)
order=np.lexsort((roll,-avg))
print(f'Best Students rolls: {order}')
```

```python
CntA=(avg>=80).sum()
CntB=((avg>=60) & (avg<80)).sum()
CntC=((avg>=40) & (avg<60)).sum()
CntD=(avg<40).sum()
```

```python
print("GRADE_COUNT")
print("A", CntA)
print("B", CntB)
print("C", CntC)
print("D", CntD)
```
