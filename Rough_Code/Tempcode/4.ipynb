```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Monthly data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# Sales data (in thousands)
product_a = np.array([23, 25, 28, 30, 35, 38, 42, 45, 48, 50, 53, 55])
product_b = np.array([30, 32, 31, 33, 35, 37, 39, 41, 43, 45, 47, 50])
product_c = np.array([15, 18, 22, 25, 28, 32, 35, 38, 42, 45, 48, 52])
```

```python
fig,axs = plt.subplots(2,2,figsize=(14,10),layout='constrained')
axs[0,0].plot(months,product_a,label="Product A")
axs[0,0].plot(months,product_b,label="Product B")
axs[0,0].plot(months,product_c,label="Product C")
axs[0,0].set_title("Monthly Sales Trends 2024")
axs[0,0].set_xlabel("Month")
axs[0,0].set_ylabel("Sales (in thousands)")
axs[0,0].legend()
axs[0,0].grid(True)


y=np.array([product_a[11],product_b[11],product_c[11]])
x=np.array(["Product A","Product B","Product C"])
bars=axs[0,1].bar(x,y)
bars[0].set_color("blue")
bars[1].set_color("green")
bars[2].set_color("red")
axs[0,1].grid(True, axis="y")
axs[0,1].set_title("December Sales Comparison")
axs[0,1].set_xlabel("Product")
axs[0,1].set_ylabel("Sales (in thousands)")


axs[1,0].scatter(months,product_a, color="blue",s=50)
axs[1,0].set_title("Product A Sales Distribution")
axs[1,0].set_xlabel("Month")
axs[1,0].set_ylabel("Sales (in thousands)")
axs[1,0].grid(True)



x=np.array(["Q1","Q2","Q3","Q4"])
y_a=product_a.reshape(4,3).sum(axis=1)
y_b=product_b.reshape(4,3).sum(axis=1)
y_c=product_c.reshape(4,3).sum(axis=1)

axs[1,1].bar(x,y_a,label="Product A",color="blue")
axs[1,1].bar(x,y_b,bottom=y_a,label="Product B",color="green")
axs[1,1].bar(x,y_c,bottom=y_a+y_b,label="Product C",color="red")

axs[1,1].set_title("Quarterly Sales (Stacked)")
axs[1,1].set_xlabel("Quarter")
axs[1,1].set_ylabel("Total Sales (in thousands)")
axs[1,1].legend()
axs[1,1].grid(True, axis="y")   # optional: y-grid looks best for bars

fig.suptitle("2024 Product Sales Analysis")
plt.tight_layout()
plt.show()
```
